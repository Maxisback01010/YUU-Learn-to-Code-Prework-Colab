### Exercise 3: Real-World List Modification (Hard)

Imagine you are managing a set of 5 numbers (for example, representing test scores). Write a program that performs the following tasks (remember: the first element is the first, the second is the second, and so on):

1. **Initialize the List:** Create a list of 5 numbers.
2. **Conditional Adjustment:** Check if the third element is less than 50.
   - If it is, update that element to 50.
3. **Conditional Penalty:** Calculate the sum of the first element and the last element.
   - If this sum is greater than 100, set the last element to 0.
4. **Swap Operation:** Swap the first element with the third element.
5. **Print the Final List:** Display the modified list.


```python
# Exercise 3, Step 1: Initialize a list of 5 numbers
#English
#create a list,put 5 numbers inside (not words this time), store it in a variable
#Pseudocode
#create a list,add 5 numbers,store it in a variable (example:scores)

# Your code here
scores=[72,45,88,60,95]
print(scores)
```

```python
# Exercise 3, Step 2: Conditional adjustment - if the third element is less than 50, update it to 50
#English
#look at the third number in the list,check is it less than 50?
#If yes-->Change it to 50
#If no--> Do nothhing
#Pseudocode
#look at the third element,if it is less than:50, change it to:50

# Your code here
if scores [2]<50:
    scores[2]=50
print(scores)
```

```python
# Exercise 3, Step 3: Conditional penalty - if the sum of the first and last elements is greater than 100, set the last element to 0
#English
#take the first,take the last number, add them together, If the total is greater than 100--> set last number to 0
#Pseudocode
#Get first element,get last element, add them, if sum>100" Change last element to 100
# Your code here
if scores [0]+scores[4]>100:
    scores[4]=0
print(scores)
```

```python
# Exercise 3, Step 4: Swap the first element with the third element
#English
#take the first element,take the third element,swap their positions
#Pseudocode
#save the first element in a temporary variable, replace the first element with the third element, replace the third element
#with the save value

# Your code here
temp=scores[0]
scores[0]=scores[2]
scores[2]=temp
print(scores)
```

```python
# Exercise 3, Step 5: Print the final list
#English
#Take your updated scores list, display it
#Pseudocode
#Show final scores list

# Your code here
print("Final scores:",scores)
```
